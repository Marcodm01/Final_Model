```python
import numpy as np
from scipy.optimize import curve_fit
from fractions import Fraction
```

```python
#Experimental values
dj1=[0,-3,12]
djr=[6,6,6]
dk=[-2,0,0]
dj=[-3,0,0]
dE=[2., 8.1045, 9.645]

v=0.28985
```

```python
def modelloNucleo(dj, a, b, g):
    dj1=[0,-3,12]
    djr=[6,6,6]
    dk=[-2,0,0]
    dEn=np.zeros(len(dj))
    for i in range(0,len(dj)):
        dEn[i]=b*dj1[i]+a*djr[i]-g*dk[i]+v*dj[i]
    return dEn
```

```python
def calcolo_abc(dj):
    dE=[2., 8.1045, 9.645]
    parametri_iniziali = [1,1,1] 
    parametri_ottimizzati, covariance = curve_fit(modelloNucleo, dj, dE, p0=parametri_iniziali)
    a_fit, b_fit, g_fit = parametri_ottimizzati
    #print(f"Parametri trovati: c = {b_fit}")
    return  a_fit, b_fit, g_fit
```

```python
def modello_random(dj1, djr, dj, dk, a, b, g):
    dE=b*dj1+a*djr+v*dj-g*dk
    return dE
```

```python
def calcolo(j1,jr,j,k):
    dj1=j1*(j1+1)-3/2*(3/2+1)
    djr=jr*(jr+1)
    dj=j*(j+1)-3/2*(3/2+1)
    dk=k**2-(3/2)**2
    return dj1, djr, dj, dk
```

```python
a=calcolo_abc(dj)[0]
b=calcolo_abc(dj)[1]
g=calcolo_abc(dj)[2]
print("a=", a)
print("b=", b)
print("g=", g)
#print("v=", v)
deltaE3=modelloNucleo(dj,a,b,g)

print("la differenza di energia tra lo stato 2 e 1 è ", deltaE3[0], "MeV, contro un valore sperimentale di 2 MeV")
print("la differenza di energia tra lo stato 3 e 1 è ", deltaE3[1], "MeV, contro un valore sperimentale di 8.104 MeV")
print("la differenza di energia tra lo stato 4 e 1 è ", deltaE3[2], "MeV, contro un valore sperimentale di 9.645 MeV")
```
