```python
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import numpy as np
```

```python
#Values
E_sper = [0, 2000, 4318.8, 4804.2, 6478.2, 8104.5 , 8420, 9645, 9780, 9970, 12510 ]
J=[3.75, 0.75, 8.75, 3.75, 15.75, 3.75,8.75, 3.75, 8.75, 15.75, 0.75]  
#E_th = [0, 2.000, 4.3188, 2.866955, 6.34775, 8.1045, 0, 9.645, 0, 0, 11.0942, 13.1232, 15.73185, 18.9292]
```

```python
etichette_sper = ['3/2₁⁻', '1/2₁⁻', '5/2₁⁻', '3/2₂⁻', '7/2₁⁻', '3/2₃⁻', '5/2₂⁻', '3/2₄⁻', '5/2₃⁻', '7/2₂⁻', '1/2₂⁻']
#etichette_th = ['3/2₁⁻', '1/2₁⁻', '5/2₁⁻', '3/2₂⁻', '7/2₁⁻', '5/2₂⁻', '3/2₃⁻', '7/2₂⁻', '3/2₄⁻', '5/2₃⁻', '9/2₁⁻', '1/2₂⁻']
```

```python
banda1=[2000, 4318.8, 6478.2]
Jbanda=[0.75, 8.75, 15.75]

m, q = np.polyfit(Jbanda, banda1, 1)
x_fit = np.linspace(0, 30, 100)
y_fit = m * x_fit + q
```

```python
fig, ax = plt.subplots()
for e,j,E in zip(etichette_sper, J, E_sper):
    ax.annotate(e, (j, E), 
                textcoords="offset points", xytext=(5, 5), 
                ha='center', fontsize=8, color='black')
    text_legend = f"{e}: {E:.1f} keV"
    ax.scatter(j, E, color='red', marker='o', label=text_legend)

plt.plot(x_fit, y_fit, color='lightgray', linestyle='--',alpha=0.7)

ax.set_xticks([0, 5, 10, 15, 20, 25, 30], ['0', '5','10', '15', '20', '25', '30'])
plt.xlim(0, 30)
plt.ylim(0, 15000)

ax.legend(loc='upper right', fontsize=8, title='Livelli Energetici')
plt.xlabel('J(J+1)')
plt.ylabel('Energia (keV)')
#plt.title('Spettro Energetico: Energie Sperimentali')
plt.grid(True, linestyle='--', alpha=0.7)
#plt.savefig('C:/Users/Marco/OneDrive/Desktop/replica.pdf', dpi=300) 
```
