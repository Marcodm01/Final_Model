```python
import numpy as np
import lmfit
```

```python
#Experimental values
dj1 = [0, -3, 12]
dl2 = [6, 6, 6]
dk = [-2, 0, 0]
dj = [-3, 0, 0]
dE = [2.0, 8.1045, 9.645]

parametri_iniziali = [1, 1, 1]
v=0.28985
```

```python
def modelloabc(dj, a, b, g):
    dE = np.zeros(len(dj))
    for i in range(len(dj)):
        dE[i] = b * dj1[i] + a * dl2[i] - g * dk[i] + v * dj[i]
    return dE
```

```python
model = lmfit.Model(modelloabc)
params = model.make_params(a=1, b=1, g=1)
result = model.fit(dE, params, dj=dj)
```

```python
a_fit = result.params['a'].value
b_fit = result.params['b'].value
g_fit = result.params['g'].value

a_err = result.params['a'].stderr
b_err = result.params['b'].stderr
g_err = result.params['g'].stderr
```

```python
print(f"a = {a_fit:.4f} ± {a_err:.4f}")
print(f"b = {b_fit:.4f} ± {b_err:.4f}")
print(f"g = {g_fit:.4f} ± {g_err:.4f}")

print(result.fit_report())
```
